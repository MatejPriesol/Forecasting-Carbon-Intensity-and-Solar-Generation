Get an idea for the distribution, plot all columns into a boxplot and histogram

```python
fig, axs = plt.subplots(3, 5, figsize=(25, 25))
axs = axs.flatten()

i = 0
for column in building3:
    ax = axs[i]
    ax.boxplot(building3[column])
    ax.title.set_text(column)
    i += 1
    
fig.suptitle("Title", fontsize=24)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

```python
bins = 20

fig, axs = plt.subplots(4, 4, figsize=(25, 25))
axs = axs.flatten()

i = 0
for column in building3:
    ax = axs[i]
    ax.hist(building3[column], bins=bins)
    ax.title.set_text(column)
    i += 1
    
fig.suptitle("Title", fontsize=24)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

```python
fig, axs = plt.subplots(5, 3, figsize=(25, 25))
axs = axs.flatten()

i = 0
for column in building3:
    ax = axs[i]
    ax.plot(building3[column])
    ax.title.set_text(column)
    i += 1
    
fig.suptitle("Title", fontsize=24)
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```
